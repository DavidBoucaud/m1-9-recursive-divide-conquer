Although merge sort has a better Big-O than selection sort, selection sort can be faster for smaller inputs.

Rewrite `merge_sort(A, min_size)` such that sub-arrays smaller than an input parameter `min_size` are sorted with our `selection_sort` from the lecture `algorithms intro`.

Time the difference between pure merge sort and this new algorithm. Is it faster? Why or why not?

````
import numpy as np

def linear_search(arr):
  current_max = arr[0]
  current_max_idx = 0
  for i in range(len(arr)):
    if arr[i] > current_max:
      current_max = arr[i]
      current_max_idx = i
  return current_max_idx

def merge(left, right):
  res = []
  while len(left)>0 and len(right)>0: 
      if left[0]<right[0]: 
          res.append(left[0]) 
          left.pop(0)
      else: 
          res.append(right[0]) 
          right.pop(0)
  for i in left: 
      res.append(i) 
  for i in right: 
      res.append(i)
  return res


def merge_selection_sort(A, min_size): 
    size = len(A)
    if size <= min_size and size > 1:
        n_sorted = 0
        while n_sorted < len(A):
            min_idx = linear_search(A[n_sorted:]) + n_sorted
            to_swap = A[n_sorted]
            A[n_sorted] = A[min_idx]
            A[min_idx] = to_swap
            n_sorted += 1

    if size > 10:
      m = size // 2
      left = merge_sort(A[m:]) 
      right = merge_sort(A[:m])
      return merge(left, right)
    else:
      return A

def merge_sort(A): 
    size = len(A)
    if size > 1:
      m = size // 2
      left = merge_sort(A[m:]) 
      right = merge_sort(A[:m])
      return merge(left, right)
    else:
      return A

list9 = [2,8,6,99,8,1,7,6,5]
list25 = [-22, 4, 5,1,900, 43, 57, 54, 387, -11, 3,2,1,4,7,6,8,9,5,7,4,0,9,8,7]

print('Merge sort on 9 elements')
%timeit merge_sort(list9)

print('Selection sort on 9 elements')
%timeit merge_selection_sort(list9, 10)

print('Merge sort on 25 elements')
%timeit merge_sort(list25)

print('Merge/Selection Hybrid on 25 elements')
%timeit merge_selection_sort(list25, 10)
````

# Exercise 2. 

Let $A[1...n]$ be an array of $n$ distinct numbers. If $i < j$ and $A[i] > A[j]$, then the pair $(i, j)$ is called an **inversion** of $A$. 

In other words an inversion is a pair of unsorted elements in an array.

**1)** List the five inversions of $[2, 3, 8, 6, 1]$ 

**2)** Give an algorithm that determines the number of inversions in any permutation on $n$ elements in $O(nlog_2(n))$ worst-case time. (Hint: Modify merge sort.)

````
# 1)
## The five inversions are (2,1), (3,1), (8,6), (8,1), (6,1).

# 2)
def Inversions(a):
    if len(a) == 1:
        return a, 0
    else:
        m = len(a) // 2
        left = a[:m]
        right = a[m:]
        left, left_ = Inversions(left)
        right, right_ = Inversions(right)
        res = []
        i = 0
        j = 0
        inversions = 0 + left_ + right_
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            res.append(left[i])
            i += 1
        else:
            res.append(right[j])
            j += 1
            inversions += (len(left)-i)
    res += left[i:]
    res += right[j:]
    return res, inversions

Inversions([2, 3, 8, 6, 1])
````

# 3. Recursive sum

Write a function that uses recursion to compute the sum of an array or list of numbers

```
recursive_sum([2, 4, 5, 6, 7])

output: 24
```

````
def recursive_sum(A):
    if len(A)==0:
        return 0
    else:
        return A[0] + recursive_sum(A[1:]) 

recursive_sum([2, 4, 5, 6, 7])
````

# 4. Recursive denominators

Write a Python program that uses recursion to find the greatest common divisor (gcd) of two integers.

```
recursive_gcd(12,14)

output : 2
```

````
def recursive_gcd(A, B):
    if B == 0:
        return A
    else:
        return recursive_gcd(B, A % B)

recursive_gcd(12,14)
````

# 5. Recursive power function

Write a recursive function to calculate the value of 'a' to the power 'b'. 

```
recursive_pow(3, 4)

output: 81
```

````
def recursive_pow(A, B):
    if B == 0:
        return 1 
    if B >= 0:
        return A * recursive_pow(A, B - 1) 

recursive_pow(6, 2)
````
